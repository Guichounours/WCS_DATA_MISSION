```python
# Séparer la colonne 'last_online' en date et heure
df_copy[['date', 'time']] = df_copy['last_online'].str.split('-', n=3, expand=True)[[0, 1]]

# Reconstruction correcte de la date et de l'heure
df_copy['date'] = df_copy['last_online'].str.slice(0, 10)  # Garde YYYY-MM-DD
df_copy['time'] = df_copy['last_online'].str.slice(11).str.replace('-', ':')  # Garde HH:MM avec format correct

# Suppression de la colonne 'last_online' si elle n'est plus utile
df_copy = df_copy.drop(columns=['last_online'])

# Vérification du résultat
print(df_copy[['date', 'time']].head())
```

```python
df_copy.info()
```

```python
# Conversion de la colonne 'date' au format datetime
df_copy['date'] = pd.to_datetime(df_copy['date'])

# Vérification de la conversion
print(df_copy['date'].dtypes)
print(df_copy['date'].head())
```

```python
# Conversion de la colonne 'time' au format horaire
df_copy['time'] = pd.to_datetime(df_copy['time'], format='%H:%M').dt.time

# Vérification de la conversion
print(df_copy['time'].dtypes)
print(df_copy['time'].head())
```
